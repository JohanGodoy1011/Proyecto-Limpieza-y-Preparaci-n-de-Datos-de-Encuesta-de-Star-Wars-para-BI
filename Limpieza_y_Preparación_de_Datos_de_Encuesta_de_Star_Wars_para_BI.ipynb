```python
# Importar librerías
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar los datos de entrada con la codificación adecuada
data = pd.read_csv("StarWars.csv", encoding="ISO-8859-1")
print(data.shape)
data.head(3)
print(["Godoy Johan"])
```

```python
data.info()
```

```python
columnas_yes_no = ['Have you seen any of the 6 films in the Star Wars franchise?',
                   'Do you consider yourself to be a fan of the Star Wars film franchise?',
                   'Are you familiar with the Expanded Universe?',
                   'Do you consider yourself to be a fan of the Expanded Universe?æ',
                   'Do you consider yourself to be a fan of the Star Trek franchise?']

# Rellenar los valores en blanco con "no"
data[columnas_yes_no] = data[columnas_yes_no].fillna('No')
data.info()
print(["Godoy Johan"])
```

```python
# Eliminar columnas no identificadas y sus preguntas relacionadas
data = data.drop(columns=[
    "Unnamed: 4", "Unnamed: 5", "Unnamed: 6", "Unnamed: 7", "Unnamed: 8", "Unnamed: 10", "Unnamed: 11", "Unnamed: 12", "Unnamed: 13",
    "Unnamed: 14", "Unnamed: 16", "Unnamed: 17", "Unnamed: 18",
    "Unnamed: 19", "Unnamed: 20", "Unnamed: 21", "Unnamed: 22", "Unnamed: 23",
    "Unnamed: 24", "Unnamed: 25", "Unnamed: 26", "Unnamed: 27", "Unnamed: 28"
])

data = data.drop(columns=["Which of the following Star Wars films have you seen? Please select all that apply.",
    "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.",
    "Please state whether you view the following characters favorably, unfavorably, or are unfamiliar with him/her.",
    "Which character shot first?"
])

# Revisar los cambios
print(data.shape)
data.info()
print(["Godoy Johan"])
```

```python
# Definir las columnas categóricas
star_wars = ['Have you seen any of the 6 films in the Star Wars franchise?',
            'Do you consider yourself to be a fan of the Star Wars film franchise?',
            'Are you familiar with the Expanded Universe?',
            'Do you consider yourself to be a fan of the Expanded Universe?æ',
            'Do you consider yourself to be a fan of the Star Trek franchise?',
            'Gender', 'Age', 'Household Income', 'Education', 'Location (Census Region)']

# Imprimir el conteo de niveles en las columnas categóricas
for col in star_wars:
    print(f'Columna {col}: {data[col].nunique()} subniveles')

data.dropna(inplace=True)
data.info()
print(["Godoy Johan"])
```

```python
# Comencemos representando la variable a predecir de forma binaria:
# 'yes' = 1, 'no' = 0
diccionario = {'Yes':1, 'No':0}
binario = data['Have you seen any of the 6 films in the Star Wars franchise?'].map(diccionario)
data['Have you seen any of the 6 films in the Star Wars franchise?_bin'] = binario

# De esta forma resulta fácil calcular la tasa de conversión: el promedio
# de la columna 'y_bin'
print(["Godoy Johan"])
```

```python
# Función para graficar tasas de conversión

def graficar_tasas_conversion(var_predictora, var_predecir, type='line', order=None):
    x, y = var_predictora, var_predecir

    # Generar agrupaciones (groupby), calcular tasa de conversión (mean),
    # multiplicarla por 100 (mul(100))
    grupo = data.groupby(x)[y].mean().mul(100).rename('tasa_conv').reset_index()

    # Y generar gráfica
    if type=='line': # Útil para rangos continuos
        plt.figure(figsize=(10,6))
        sns.lineplot(x=var_predictora, y='tasa_conv', data=grupo)
        plt.grid()
    elif type=='bar': # Útil si los datos están divididos en rangos o son categóricos
        plt.figure(figsize=(14,6))
        sns.barplot(x=var_predictora, y='tasa_conv', data=grupo, order=order)
        plt.grid()
    elif type=='scatter': # Útil si los datos están divididos en rangos o son categóricos
        plt.figure(figsize=(10,6))
        sns.scatterplot(x=var_predictora, y='tasa_conv', data=grupo)
        plt.grid()
print(["Godoy Johan"])
```

```python
# Y grafiquemos la tasa de conversión para esta nueva columna
graficar_tasas_conversion('Do you consider yourself to be a fan of the Star Wars film franchise?','Have you seen any of the 6 films in the Star Wars franchise?_bin',type='bar')
print(["Godoy Johan"])
```

```python
# Tasa de conversión para dos columnas
def graficar_tc_bivariada(col_list, orden=None):
    tc = data.groupby(col_list)['Have you seen any of the 6 films in the Star Wars franchise?_bin'].mean().to_frame().reset_index()

    # Gráfica
    plt.figure(figsize=(13, 7))
    ax = sns.pointplot(x=tc['Have you seen any of the 6 films in the Star Wars franchise?_bin'], y=tc[col_list[0]], hue=tc[col_list[1]], join=False, order=orden)
    ax.yaxis.grid(True)
    ax.xaxis.grid(True)
    plt.title(f'Tasa de conversión para {col_list[0]} y {col_list[1]}')
    plt.xlabel('Tasa de conversión (%)')
    plt.xlim((0, 1))
graficar_tc_bivariada(['Age', 'Gender'])
print(["Godoy Johan"])
```
